```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
# Using indices as x-axis
y = np.array([1,4,5,14,23])**2 
plt.plot(y)
plt.show()
```

```python
x = np.arange(10)
y = x**2
plt.figure(figsize=(10,10)) # Adjusting the size of the figure
plt.scatter(x,y,label = 'kiwi',color = 'black',marker = '^')
plt.xlabel('time')
plt.ylabel('prices')
plt.legend()
plt.show()
```
