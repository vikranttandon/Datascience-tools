```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To see what all themes are available
themes = plt.style.available
print(themes)
```

```python
# Now choose one and pass it in 
plt.style.use('dark_background')
```

```python
# Constructing line plots for 2 functions, 
'''
y = f(x): mx + c
y = g(x): x^2

'''
def f(m,x,c): # linear
    return m*x + c
def g(x): # exponential
    return x**2

# Receiving multiple points on the line for both the lines in arrays

linear_x = np.array([2,3,5,7,8,11,14])

m = 7
c = 1
linear_y = np.array([f(m,cur,c) for cur in linear_x])
linear_y = linear_y - linear_x # profit 

exp_x = np.array([2,4,7,9,11,14])
exp_y = np.array([g(cur) for cur in exp_x])
exp_y = exp_y - exp_x # profit
# Managed together

plt.plot(linear_x,linear_y,color = 'red',label = 'uber_bike',marker = 'o')
plt.plot(exp_x,exp_y,color = 'blue',label = 'uber_car',marker = '*')
 # To give title and labels
plt.title('profit curves')
plt.xlabel('investment')
plt.ylabel('return')
plt.legend() # To reflect the labeling of the curves
plt.show()

# Managed separately 
plt.plot(linear_x,linear_y,color = 'green')
plt.show()

plt.plot(exp_x,exp_y,color = 'yellow') 
plt.show()
```
