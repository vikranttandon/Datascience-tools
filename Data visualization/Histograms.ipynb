```python
Xsn = np.random.randn(100)
```

Since, Xsn = (X-mean)/std-deviation 
<br>=> <font color = 'red'>X = (Xsn x std-deviation) + mean

```python
u = 60
sigma = 5
X = (Xsn*sigma) + u
```

```python
plt.style.use('seaborn')
```

```python
plt.hist(X,edgecolor = 'black')
plt.ylabel('prob/ frequency of students')
plt.xlabel('Marks Range')

plt.ylim((0,25))
plt.show()
```

```python
# For plotting 2 graphs, overlapping can be taken care of alpha(manages transparency of histogram).
phy = np.random.randn(100)
maths = np.random.randn(100)

u_p = 40
sig_p = 11

u_m = 60
sig_m = 17

X_phy = phy*sig_p + u_p
X_maths = maths*sig_m + u_m

plt.hist(X_phy,edgecolor = 'black',label='Physics')
plt.hist(X_maths,edgecolor = 'black',label = 'Maths',alpha = 0.8)

plt.xlabel('Marks Range')
plt.ylabel('Prob')

plt.legend()
plt.show()
```
